```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

def simple_linear_regression(x, y):
    # Add a constant to the independent variable
    X = sm.add_constant(x)

    # Fit the model
    model = sm.OLS(y, X)
    results = model.fit()

    # Print the summary statistics
    print(results.summary2(float_format="%.4f"))

    # Calculate the residuals
    residuals = results.resid

    # Create a new DataFrame for the results
    results_df = pd.DataFrame({
        'x': x,
        'y': y,
        'a + bx': results.fittedvalues,
        'Residual Error': residuals
    })

    return results_df

# Load the data
data = pd.read_excel('iris_dataset.xlsx', header=1)

# Filter the data for the 'setosa' species
data = data[data['Species'] == 'setosa']

# Define the dependent variable (y) and independent variable (x)
y = data['Sepal.Width']
x = data['Sepal.Length']

# Perform the regression and get the results
results = simple_linear_regression(x, y)

# Display the results
print("\nTable with residuals and predicted values:")
display(results)
```
